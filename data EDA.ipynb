```python
# 📦 Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 🔍 Read the Excel file
file_path = '数据_备用.xlsx'
df = pd.read_excel(file_path)

# ⚡️ 1. Basic Overview of the Data
print("Shape of the dataset:", df.shape)
print("\nFirst 5 rows:")
print(df.head())   # Display first 5 rows
print("\nData Summary:")
print(df.info())   # Information about data types and missing values
print("\nStatistical Summary:")
print(df.describe())   # Basic statistical description of numerical columns

# 🚩 2. Missing Values Check
print("\nMissing Values in Each Column:")
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0])   # Only show columns with missing values

# 🌡️ 3. Heatmap for Missing Data Visualization
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

# 📊 4. Correlation Matrix
plt.figure(figsize=(12, 8))
corr_matrix = df.corr()   # Correlation between numerical variables
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

```

```python
# ✅ Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=[np.number])  # This filters only numerical columns

# 📊 Correlation Matrix (Fixed)
plt.figure(figsize=(12, 8))
corr_matrix = numeric_df.corr()   # Now it's safe to compute correlation
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix for Numeric Variables')
plt.show()
```

```python
# 🎯 仅显示相关系数大于0.3或小于-0.3的部分
filtered_corr = corr_matrix[(corr_matrix >= 0.3) | (corr_matrix <= -0.3)]

plt.figure(figsize=(16, 12))
sns.heatmap(
    filtered_corr, 
    annot=True, 
    cmap='coolwarm', 
    fmt=".2f", 
    linewidths=0.5,
    annot_kws={"size": 12}
)
plt.title('Filtered Correlation Matrix (|r| > 0.3)', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# 📈 5. Distribution of Key Variables (Histogram)
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()  # Select numeric columns
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.show()

```

```python
# 📦 6. Boxplot to Detect Outliers
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()


```

```python
# 🔗 7. Scatter Plot for Key Relationships
# Example: Relationship between Same_Surname and CAR (if applicable)
if 'Same_Surname' in df.columns and 'Buyer_CAR1' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x='Same_Surname', y='Buyer_CAR1', data=df)
    plt.title('Same Surname vs CAR')
    plt.show()

```

```python
# 🔗 7. Scatter Plot for Key Relationships
# Example: Relationship between Same_Surname and CAR (if applicable)
if 'Same_Surname' in df.columns and 'Seller_CAR1' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x='Same_Surname', y='Seller_CAR1', data=df)
    plt.title('Same Surname vs CAR')
    plt.show()
```

```python
# 🧮 8. Grouped Analysis (e.g., Mean CAR by Same_Surname)
if 'Same_Surname' in df.columns and 'Buyer_CAR1' in df.columns:
    grouped_data = df.groupby('Same_Surname')['Buyer_CAR1'].mean().reset_index()
    print("\nAverage CAR by Same Surname:")
    print(grouped_data)

    plt.figure(figsize=(6, 4))
    sns.barplot(x='Same_Surname', y='Buyer_CAR1', data=grouped_data, palette='Set2')
    plt.title('Average Buyer_CAR1 for Same Surname vs Non-Same Surname')
    plt.show()
```

```python
# 🧮 8. Grouped Analysis (e.g., Mean CAR by Same_Surname)
if 'Same_Surname' in df.columns and 'Seller_CAR1' in df.columns:
    grouped_data = df.groupby('Same_Surname')['Seller_CAR1'].mean().reset_index()
    print("\nAverage CAR by Same Surname:")
    print(grouped_data)

    plt.figure(figsize=(6, 4))
    sns.barplot(x='Same_Surname', y='Seller_CAR1', data=grouped_data, palette='Set2')
    plt.title('Average Seller_CAR1 for Same Surname vs Non-Same Surname')
    plt.show()
```
